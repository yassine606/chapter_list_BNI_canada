```python
from selenium import webdriver
import time 
from selenium.webdriver.support.select import Select
import pandas as pd 
```

```python
driver=webdriver.Chrome(executable_path='/Users/admin/Desktop/chromedriver')
driver.get('https://bnicanada.ca/en-CA/chapterlist?countryIds=17114&regionId=&chapterName=&chapterCity=&chapterArea=&chapterMeetingDay=&chapterMeetingTime=&chapterMeetingType=')
driver.maximize_window()
time.sleep(1)

```

```python
#pour les cookies
cookie=driver.find_element_by_xpath('//a[text()="I understand"]')
try : 
    cookie.click()

except :
    pass
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
col_names =  ["Name"]
df = pd.DataFrame(columns = col_names)
chapter_regions=[]
urls=[]
cities=[]
areas=[]
Meeting_days=[]
Meeting_times=[]
meeting_types=[]

for _ in range(0,5):
    
    chapter_region = driver.find_elements_by_xpath('//tr[@role="row"]/td[1]/a')
    url = driver.find_elements_by_xpath('//tr[@role="row"]/td[1]/a')
    city = driver.find_elements_by_xpath("//tr[@role='row']/td[2]")
    area = driver.find_elements_by_xpath("//tr[@role='row']/td[3]")
    Meeting_day = driver.find_elements_by_xpath("//tr[@role='row']/td[4]")
    Meeting_time = driver.find_elements_by_xpath("//tr[@role='row']/td[5]")
    meeting_type = driver.find_elements_by_xpath("//tr[@role='row']/td[6]")

    for i in range(0,len(chapter_region)):
        chapter_regions.append(chapter_region[i].text)
        urls.append(url[i].get_attribute("href"))
        cities.append(city[i].text)
        areas.append(area[i].text)
        Meeting_days.append(Meeting_day[i].text)
        Meeting_times.append(Meeting_time[i].text)
        meeting_types.append(meeting_type[i].text)
    try:
        next_button = driver.find_element_by_xpath("//i[@class='fa fa-chevron-right']")
        next_button.click()
        wait = WebDriverWait(driver, 10)
    except:
        break
        
        
df["chapter_region"], df["url"],df["city"], df["area"], df["Meeting_day"],df['Meeting_times'],df['meeting_types']  = chapter_regions, urls, cities, areas, Meeting_days,Meeting_times,meeting_types


```
